## Problem Definition

- Date represents weekly retail scan data for Nation retail volume (units) & price
- Retail scan data comes directly from retailers' cash registers based on actual retail sales of of Hass avocados
- Starting in 2013, the table below reflects an expanded, multi-outlet retail data set. Multi-outlet reporting includes an aggregation of the following channels: grocery, club, mass, drug, dollar, & military
- The Average Price (of avocados) in the table reflects a per unit (per avocado) cost, even when multiple units (avocados) are sold in bags
- The Product lookup codes (PLU's) in the table are only for Hass avocadoes. Other varieties of avocadoes (e.g. greenskins) are not included in this table

Relevant columns in the dataset:

    - Date - the date of the observation
    - Average Price - the average price of a single avocado
    - Type - conventional or organic
    - Year - The year
    - Region - The city or the region of the observation
    - Total Volume - Total # of avocados sold
    - 4046 - Total # of avocados with PLU 4046 sold
    - 4225 - Total # of avocados with PLU 4225 sold
    - 4770 - Total # of avocados with PLU 4770 sold

## Make Predictions

```python
# In order to run Prophet the columns names must be 'ds' & 'y'
avocado_prophet_df = avocado_prophet_df.rename(columns={'Date': 'ds', 'AveragePrice': 'y'})
```

```python
avocado_prophet_df
```

```python
m = Prophet()
```

```python
# Fit the model
m.fit(avocado_prophet_df)
```

```python
# Forecast the Future
future = m.make_future_dataframe(periods = 365)
forecast = m.predict(future)
```

```python
forecast
```

```python
figure = m.plot(forecast, xlabel = 'Date', ylabel = 'Price')
```

```python
figure = m.plot_components(forecast)
```
